1.Product


```python
from itertools import product
```

```python
a=[1,2]
b=[3,4]

prod=product(a,b) #gives us combination of elements
print(list(prod))
```

2.permutations  ---gives all permutations of elements

```python
from itertools import permutations
```

```python
x=[1,2,3,4]
per=permutations(x)
per1=permutations(x,2) #second argument is length of element
```

```python
#print(list(per))
print(list(per1))
```

3.combinations --make all the combinations of elements (permutation vs combination in combination same pair does not repeat

```python
from itertools import combinations ,combinations_with_replacement
```

```python
y=[1,2,3,4]
com=combinations(y,2)
com1=combinations_with_replacement(y,2) #also makes similar element  combinations 
```

```python
print(list(com))
print(list(com1))
```

4.accumulate:-  accumulates sum of subsequent elements

```python
from itertools import accumulate
import operator
```

```python
w=[1,2,3,4]
acc =accumulate(w) #sum of values
acc_mul =accumulate(w,func=operator.mul)#multiplication  of values
print(w)
print(list(acc))
print(list(acc_mul))
```

```python
w1=[1,2,5,6,3,4]
acc_max =accumulate(w1,func=max)#find max  of values
print(list(acc_max))
```

5.Groupby :-  seperates elements in different group according to key

```python
from itertools import groupby
```

```python
def smaller_thn_3(x):
    return x<3
```

```python
p=[1,2,3,4,5,6]
grp =groupby(p,key=smaller_thn_3)

for key,value in grp:
    print(key,list(value))
```

```python
for i in  count(10):  #count infinite from input value
    print(i)
    if i==15:
        break
```
